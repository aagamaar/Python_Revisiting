LISTS - ordered and changeable (mutable)

```python
fruits = ["apple", "mango", "orange", "banana"]
print(fruits)
print(fruits[0])
print(fruits[2])

#changing item
fruits[1]="strawberry"
print(fruits)

print(type(fruits))
```

```python
fruits.append("grapes")
print(fruits)
```

```python
fruits.remove("banana")
print(fruits)
```

```python
print("Number of fruits: ", len(fruits))
```

```python
for fruit in fruits:
  print (fruit)
```
