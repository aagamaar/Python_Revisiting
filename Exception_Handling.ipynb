Try and Except

```python
try:
  num = int(input("Enter a number: "))
  print("Your number is", num)
except ValueError:
  print("That was not a integer")
```

```python
try:
  num = int(input("Enter a number: "))
  print("Your number divided by 10 is", num/10)
except ZeroDivisionError:
  print("Type a valid number :)")
```

Catching any errors

```python
try:
  num = int(input("Enter a number: "))
  print("10 divided by Your number is", 10/num)
except:
  print("Type a valid number :)")
```

```python
try:
  num = int(input("Enter a number: "))
  print("10 divided by Your number is", 10/num)
except Exception as e:
  print("Type a valid number :)", str(e))
```

```python
try:
  num = int(input("Enter a number: "))
  r=10/num
  print("10 divided by Your number is", r)
except ValueError:
  print("Type a valid number :)")
finally:
  print("Finally block runs no matter what")
```
