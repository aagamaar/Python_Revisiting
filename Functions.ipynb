To add 2 numbers

```python
def add(a,b):
  sum = a+b
  print("The sum of",a, " & ",b, "is",sum)
```

**Square of numbers**

```python
def square(n):
  sq=n*n
  result = sq
  print("Square is",result)
```

```python
square(9)
```
