Dictionary - key value pairs


```python
person = {
    "name":"Alice",
    "age":21,
    "is_student":True
}

print(person)
```

```python
person["age"]
```

```python
person["age"] = 23
```

```python
print(person)
```

```python
person["city"] = "Kochi"
print(person)
```

```python
for key, value in person.items():
  print(key, "-->",value)
```
