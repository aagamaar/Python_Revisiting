For loop

```python
print(*range(5))
```

```python
print(*range(2,9))
```

```python
for i in range(5):
  print("Hello")
```

```python
for i in range(0,10):
  print(i)
```

```python
for x in range(1,6):
  print( x , x*x)
```

```python
count = 1

while count <=5 :
  print("Hello")
  count+=1
```

```python

# ❌ Be careful: this loop never stops
while True:
  print("This is an infinite loop")
```

```python
for i in range(1,6):
  if i==3:
    break
  print(i)
```

```python
for i in range(1,6):
  if i==3:
    continue
  print(i)
```
